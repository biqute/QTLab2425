```python
import sys; sys.path.append("../strumenti") 
sys.path.append("../utils")
from VNA import VNA
import matplotlib.pyplot as plt
import numpy as np
from utils import Data


ip = '192.168.40.10'

vna = VNA(ip)
```

```python
vna.set_freq_span(6.2358, 0.003)
vna.set_sweep_points(1001)
vna.set_n_means(10)
vna.set_ifband(1000)
vna.set_power(-40)
```

```python
phi = vna.get_phase()
freq = vna.get_freq()
powe = vna.get_dbm()
I, Q = vna.get_S_parameters("S21")
data = Data()
data.plot(freq, powe)
data.plot(freq, phi)
```
