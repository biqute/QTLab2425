```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import iminuit as minuit
from iminuit import Minuit
from iminuit.cost import LeastSquares
sys.path.append("../../fit")
import resonance_fit
```

# TERZA RISONANZA 1.4K AL VARIARE DI V_bias

```python
y = []
V = [0.0, 0.25, 0.5, 0.8, 1.0, 1.25, 1.5, 1.8, 2.0, 2.25, 2.5, 2.8, 3.0, 3.25, 3.5, 3.8]

fr_list = []
Qc_list = []
Qt_list = []
a_list = []
b_list = []
c_list = []
d_list = []
k_list = []
phi_list = []

initial_params = {
    'a': 1,  # Ultimo valore di fr ottenuto
    'b': 10e-9,  # Ultimo valore di Qc ottenuto
    'c': 10e-18,       # Impostazione predefinita per c, d, k, phi, Qt, Qc, fr
    'd': 10e-27,
    'k': 2.5,
    'phi': 0.42,
    'Qt': 10000,
    'Qc': 75000,
    'fr': 5809178611  # Ultima frequenza di risonanza come valore iniziale
}

for volt in V:
    file_path = f"Cavity_resonances_1.4K/Resonance3/res3-50dBm_{volt}V.txt"
    fr, qc, qt, a, b, c, d, k, phi = resonance_fit.fit(file_path, initial_params=initial_params)

    fr_list.append(fr)
    Qc_list.append(qc)
    Qt_list.append(qt)
    a_list.append(a)
    b_list.append(b)
    c_list.append(c)
    d_list.append(d)
    k_list.append(k)
    phi_list.append(phi)

    # Usa i parametri ottenuti dal fit precedente come parametri iniziali per il prossimo fit
    initial_params = {
        'a': 1,  # Ultimo valore di fr ottenuto
        'b': 10e-9,  # Ultimo valore di Qc ottenuto
        'c': 10e-18,       # Impostazione predefinita per c, d, k, phi, Qt, Qc, fr
        'd': 10e-27,
        'k': 3,
        'phi': 0.292,
        'Qt': Qt_list[-1],
        'Qc': Qc_list[-1],
        'fr': fr_list[-1]  # Ultima frequenza di risonanza come valore iniziale
    }


f_r = np.array(fr_list)
y=(f_r-f_r[0])/ f_r[0]  # Calcola la variazione percentuale della frequenza di risonanza rispetto al primo valore
V = np.array(V)


x = V / 1.997
```

Salvataggio parametri fit

```python
import os

# Percorso relativo (la cartella verr creata se non esiste)
output_dir = "Cavity_resonances_1.4K/Resonance3"
os.makedirs(output_dir, exist_ok=True)

# Componi il percorso completo al file
output_file = os.path.join(output_dir, "fit_results.txt")

# Salvataggio del file
with open(output_file, 'w') as f:
    f.write(f"{'V (V)':>6} {'fr (Hz)':>15} {'Qc':>10} {'Qt':>10} {'a':>6} {'b':>10} {'c':>10} {'d':>10} {'k':>6} {'phi':>8}\n")
    f.write("-" * 100 + "\n")
    for i in range(len(V)):
        f.write(f"{V[i]:6.2f} {fr_list[i]:15.6e} {Qc_list[i]:10.2f} {Qt_list[i]:10.2f} "
                f"{a_list[i]:6.2f} {b_list[i]:10.2e} {c_list[i]:10.2e} {d_list[i]:10.2e} "
                f"{k_list[i]:6.2f} {phi_list[i]:8.4f}\n")

print(f"Risultati salvati in '{output_file}'")
```

fit I bias

```python
def func (x, I):
    return -1/2*(x**2/I**2)

y_err = np.ones(len(y))*1e-4

# Define the cost function (Least Squares)
least_squares = LeastSquares(x, y, y_err, func)

# Initialize Minuit with the cost function and initial parameter guess
m = Minuit(least_squares, I=1.0)

# Perform the fit
m.migrad()  # Minimize the cost function
m.hesse()   # Estimate uncertainties

plt.errorbar(x, y, yerr=y_err, fmt='o', label='Data', color='blue')
plt.plot(x, func(x,m.values["I"]), label=f'Fit: I = {m.values["I"]:.3f}  {m.errors["I"]:.3f}', color='Blue')

# Add labels, legend, and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fit of func(x, I) to data')
plt.legend()
plt.grid()

# Show the plot
plt.show()
```
