# FIT I_bias: RES 3

```python
import numpy as np
import matplotlib.pyplot as plt
from iminuit import Minuit
from iminuit.cost import LeastSquares

x = V/1.997
y_err = np.ones_like(y) * 1e-3
# ---- Funzioni ----

# Funzione fisica esatta
def exact_func(x, Ibias):
    return 1 / np.sqrt(1 + (x / Ibias)**2) - 1

# Espansione di Taylor alternativa (Ibias/x)^2
def taylor_exp(x, Ibias):
    return -0.5 * (x / Ibias)**2

# ---- Fit 1: funzione esatta ----
least_squares1 = LeastSquares(x, y, y_err, exact_func)
m1 = Minuit(least_squares1, Ibias=5)
m1.migrad()
m1.hesse()
Ibias_exact = m1.values["Ibias"]
Ibias_exact_err = m1.errors["Ibias"]

chi2_1 = least_squares1(m1.values)  # restituisce 
ndof1 = len(x) - m1.nfit
chi2r_1 = chi2_1 / ndof1


# ---- Fit 2: funzione Taylor ----
least_squares2 = LeastSquares(x, y, y_err, taylor_exp)
m2 = Minuit(least_squares2, Ibias=5)
m2.migrad()
m2.hesse()
Ibias_taylor = m2.values["Ibias"]
Ibias_taylor_err = m2.errors["Ibias"]

chi2_2 = least_squares2(m2.values)
ndof2 = len(x) - m2.nfit
chi2r_2 = chi2_2 / ndof2

print("=== Fit Esatto ===")
print(f"Ibias = {Ibias_exact:.5f}  {Ibias_exact_err:.5f}")
print(f" = {chi2_1:.2e}, dof = {ndof1}, _r = {chi2r_1:.3f}\n")

print("=== Fit Taylor ===")
print(f"Ibias = {Ibias_taylor:.5f}  {Ibias_taylor_err:.5f}")
print(f" = {chi2_2:.2e}, dof = {ndof2}, _r = {chi2r_2:.3f}")
# ---- Plot ----
plt.errorbar(x, y, yerr=y_err, fmt='o', label='Dati', color='black')

# Fit con funzione esatta
x_fit = np.linspace(np.min(x), np.max(x), 500)
plt.plot(x_fit, exact_func(x_fit, Ibias_exact), 'b-',
         label=f'Fit esatto: Ibias = {Ibias_exact:.3f}  {Ibias_exact_err:.3f}')

# Fit con funzione Taylor
plt.plot(x_fit, taylor_exp(x_fit, Ibias_taylor), 'r--',
         label=f'Fit Taylor: Ibias = {Ibias_taylor:.3f}  {Ibias_taylor_err:.3f}')

# --- Etichette ---
plt.xlabel("Corrente (x)")
plt.ylabel("f / f (y)")
plt.title("Confronto Fit: funzione esatta vs Taylor")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# FIT I_bias: RES 4

```python
import numpy as np
import matplotlib.pyplot as plt
from iminuit import Minuit
from iminuit.cost import LeastSquares


x = V/1.997
y_err = np.ones_like(y) * 1e-3

# Funzione esatta
def exact_func(x, Ibias):
    return 1 / np.sqrt(1 + (x / Ibias)**2) - 1

# Espansione di Taylor alternativa
def taylor_exp(x, Ibias):
    return -0.5 * (x / Ibias)**2

# ---- Fit 1: funzione esatta ----
least_squares1 = LeastSquares(x, y, y_err, exact_func)
m1 = Minuit(least_squares1, Ibias=10)
m1.migrad()
m1.hesse()
Ibias_exact = m1.values["Ibias"]
Ibias_exact_err = m1.errors["Ibias"]

chi2_1 = least_squares1(m1.values) 
ndof1 = len(x) - m1.nfit
chi2r_1 = chi2_1 / ndof1


# ---- Fit 2: funzione Taylor ----
least_squares2 = LeastSquares(x, y, y_err, taylor_exp)
m2 = Minuit(least_squares2, Ibias=10)
m2.migrad()
m2.hesse()
Ibias_taylor = m2.values["Ibias"]
Ibias_taylor_err = m2.errors["Ibias"]

chi2_2 = least_squares2(m2.values)
ndof2 = len(x) - m2.nfit
chi2r_2 = chi2_2 / ndof2

print("=== Fit Esatto ===")
print(f"Ibias = {Ibias_exact:.5f}  {Ibias_exact_err:.5f}")
print(f" = {chi2_1:.2e}, dof = {ndof1}, _r = {chi2r_1:.3f}\n")

print("=== Fit Taylor ===")
print(f"Ibias = {Ibias_taylor:.5f}  {Ibias_taylor_err:.5f}")
print(f" = {chi2_2:.2e}, dof = {ndof2}, _r = {chi2r_2:.3f}")
# ---- Plot ----
plt.errorbar(x, y, yerr=y_err, fmt='o', label='Dati', color='black')

# Fit con funzione esatta
x_fit = np.linspace(np.min(x), np.max(x), 500)
plt.plot(x_fit, exact_func(x_fit, Ibias_exact), 'b-',
         label=f'Fit esatto: Ibias = {Ibias_exact:.3f}  {Ibias_exact_err:.3f}')

# Fit con funzione Taylor
plt.plot(x_fit, taylor_exp(x_fit, Ibias_taylor), 'r--',
         label=f'Fit Taylor: Ibias = {Ibias_taylor:.3f}  {Ibias_taylor_err:.3f}')

plt.xlabel("Corrente (x)")
plt.ylabel("f / f (y)")
plt.title("Confronto Fit: funzione esatta vs Taylor")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
