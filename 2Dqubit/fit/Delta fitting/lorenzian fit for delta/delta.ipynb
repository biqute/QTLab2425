# FIT RISONANZE AL VARIARE DI T PER LA DETERMINAZIONE DEL GAP SUPERCONDUTTIVO

## PRIMO RISONATORE (file Qres pari)

```python
import numpy as np
import matplotlib.pyplot as plt
import sys 
sys.path.append("../../../fit")
sys.path.append("../../../data")
import resonance_fit
```

Preparazione array temperature e files

```python
files = []
p = np.arange (40, 118, 2)
Temp = np.arange (30, 420, 10)
Temp = Temp[::-1]
j = 0
for i in p :
    files.append (f"Q_res{i}")
    print(f"file = Q_res{i} a temperatura T = {Temp[j]}")
    j = j +1





```

fitting delle risonanza

```python

fr_list = []
Qc_list = []
Qt_list = []
Qi_list = []
a_list = []
b_list = []
c_list = []
d_list = []
k_list = []
phi_list = []

initial_params = {
    'a': 1,  # Ultimo valore di fr ottenuto
    'b': 10e-9,  # Ultimo valore di Qc ottenuto
    'c': 10e-18,       # Impostazione predefinita per c, d, k, phi, Qt, Qc, fr
    'd': 10e-27,
    'k': 2.5,
    'phi': 0.04,
    'Qt': 10003,
    'Qc': 20000,
    'fr': 317800000  # Ultima frequenza di risonanza come valore iniziale
}

for elem, T in zip(files, Temp):
    file_path = f"../../../data/Delta/gap_run12/{elem}.txt"
    
    # Fit del file
    fr, qc, qt, a, b, c, d, k, phi = resonance_fit.fit_cut(file_path, None, "," , initial_params=initial_params)

    # Salva temperatura e risultati
    qi = 1/(1/qt - 1/qc)
    fr_list.append(fr)
    Qc_list.append(qc)
    Qt_list.append(qt)
    Qi_list.append(qi)
    a_list.append(a)
    b_list.append(b)
    c_list.append(c)
    d_list.append(d)
    k_list.append(k)
    phi_list.append(phi)

    # Parametri iniziali per il fit successivo
    initial_params = {
        'a': 1,
        'b': 1e-8,
        'c': 1e-17,
        'd': 1e-26,
        'k': 1,
        'phi': 0.01,
        'Qt': qt,
        'Qc': qc,
        'fr': fr
    }

    print(f"T = {T}, Qi = {qi}")

```

```python
Qt = np.array(Qt_list)
Qc = np.array(Qc_list)
Qi = np.array(Qi_list)
inv_Qi = 1/Qi
params = [Qc_list, Qt_list, fr_list, Qi, inv_Qi]  # Lista dei parametri da plottare
param_names = ['qc', 'qt', 'fr', 'Qi', 'inv_Qi']  # Nome dei parametri

# Iteriamo su parametri e nomi
for elem, param_name in zip(params, param_names):
    fig, ax = plt.subplots(1, 1)
    ax.plot(Temp, elem)
    ax.set_title(f"Grafico di {param_name}")
    ax.set_xlabel("Temperatura (mK)")  # Assumendo che 'Temp' sia in mK
    ax.set_ylabel(f"{param_name} Value")  # Etichetta personalizzata per ogni parametro
    plt.show()
```
