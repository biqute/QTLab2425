## DELTA FIT WITH IQ CIRCLE METHOD

```python
from circle_for_delta import fit_full_resonator
import numpy as np
import matplotlib.pyplot as plt
files = []
qc_list = []
qi_list = []
qt_list = []
fr_list = []
p = np.arange (40, 118, 2)
Temp = np.arange (30, 420, 10)
Temp = Temp[::-1]
j = 0
for i in p :
    files.append (f"Q_res{i}")
    print(f"file = Q_res{i} a temperatura T = {Temp[j]}")
    j = j +1

for elem, T in zip(files, Temp):
    file_path = f"../../data/Delta/gap_run12/{elem}.txt"
    qc, qi, qt, fr = fit_full_resonator(file_path)
    qc_list.append(qc)
    qi_list.append(qi)
    qt_list.append(qt)
    fr_list.append(fr)
    
Qt = np.array(qt_list)
Qc = np.array(qc_list)
Qi = np.array(qi_list)

inv_Qi = 1/Qi
params = [qc_list, qt_list, Qi, inv_Qi] 
param_names = ['qc', 'qt', 'Qi', 'inv_Qi']  


for elem, param_name in zip(params, param_names):
    fig, ax = plt.subplots(1, 1)
    ax.plot(Temp, elem)
    ax.set_title(f"Grafico di {param_name}")
    ax.set_xlabel("Temperatura (mK)")  
    ax.set_ylabel(f"{param_name} Value")  
    plt.show()
```

```python
from circle_for_delta import fit_delta
Qc_cut = []
Qt_cut = []
Qi_cut = []
T_cut = []
T_err = {60} 

for qi, temp in zip(qi_list, Temp):
    if temp not in T_err: 
        Qi_cut.append(qi)
        T_cut.append(temp)
        

cut = T_cut.index(30)
T_fit = T_cut[:cut + 1]  
Qi_fit = Qi_cut[:cut + 1]

fit_delta(Qi_fit, T_fit, freq=fr_list[0], alpha_fixed=0.8)
```
