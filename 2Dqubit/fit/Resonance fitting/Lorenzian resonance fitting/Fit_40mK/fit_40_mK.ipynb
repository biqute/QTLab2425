Spostamento della frequenza di risonanza

```python
fr = inverted_list = [fr_list[i] for i in range(len(fr_list)-1, -1, -1)]
dBm = [-35, -40, -45, -50, -55, -60]

plt.figure(figsize=(8, 5))

plt.plot(fr, dBm, color='lightgray', linewidth=2)
plt.scatter(fr, dBm, color='green', s=50, zorder=3)
plt.xlabel('frequency')
plt.ylabel('power (dBm)')
plt.title('power vs frequency')
plt.grid(True)

plt.show()
```
