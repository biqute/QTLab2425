# FIT I_bias: RES 3

```python

x = V/1.997
y_err = np.ones_like(fr) * 1e-3


def taylor_exp(x, Ibias):
    return -0.5 * (x / Ibias)**2


least_squares2 = LeastSquares(x, y, y_err, taylor_exp)
m2 = Minuit(least_squares2, Ibias=5)
m2.migrad()
m2.hesse()
Ibias_taylor = m2.values["Ibias"]
Ibias_taylor_err = m2.errors["Ibias"]

chi2_2 = least_squares2(m2.values)
ndof2 = len(x) - m2.nfit
chi2r_2 = chi2_2 / ndof2

print("=== Fit Taylor ===")
print(f"Ibias = {Ibias_taylor:.5f} ± {Ibias_taylor_err:.5f}")
print(f"χ² = {chi2_2:.2e}, dof = {ndof2}, χ²_r = {chi2r_2:.3f}")
# ---- Plot ----
plt.errorbar(x, y, yerr=y_err, fmt='o', label='Dati', color='black')

x_fit = np.linspace(np.min(x), np.max(x), 500)

plt.plot(x_fit, taylor_exp(x_fit, Ibias_taylor), 'r--',
         label=f'Fit Taylor: Ibias = {Ibias_taylor:.3f} ± {Ibias_taylor_err:.3f}')

plt.xlabel("Corrente (x)")
plt.ylabel("Δf / f₀ (y)")
plt.title("Confronto Fit: funzione esatta vs Taylor")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# FIT I_bias: RES 4

```python


x = V/1.997
y_err = np.ones_like(fr) * 1e-3


def taylor_exp(x, Ibias):
    return -0.5 * (x / Ibias)**2




least_squares2 = LeastSquares(x, y, y_err, taylor_exp)
m2 = Minuit(least_squares2, Ibias=10)
m2.migrad()
m2.hesse()
Ibias_taylor = m2.values["Ibias"]
Ibias_taylor_err = m2.errors["Ibias"]

chi2_2 = least_squares2(m2.values)
ndof2 = len(x) - m2.nfit
chi2r_2 = chi2_2 / ndof2

print("=== Fit Taylor ===")
print(f"Ibias = {Ibias_taylor:.5f} ± {Ibias_taylor_err:.5f}")
print(f"χ² = {chi2_2:.2e}, dof = {ndof2}, χ²_r = {chi2r_2:.3f}")
# ---- Plot ----
plt.errorbar(x, y, yerr=y_err, fmt='o', label='Dati', color='black')

x_fit = np.linspace(np.min(x), np.max(x), 500)


plt.plot(x_fit, taylor_exp(x_fit, Ibias_taylor), 'r--',
         label=f'Fit Taylor: Ibias = {Ibias_taylor:.3f} ± {Ibias_taylor_err:.3f}')

plt.xlabel("Corrente (x)")
plt.ylabel("Δf / f₀ (y)")
plt.title("Confronto Fit: funzione esatta vs Taylor")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
