# Here we cut data to accurately integrate the square signal

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read raw data
frequency = 6e3
duty_cycle = 99
with h5py.File('../Experiments/PhotodiodeArea/run-1/raw_data.h5', 'r') as f:
    raw_data = f[f'freq-{frequency}_duty-{duty_cycle}'][:]
    
voltage = raw_data[1]
time = raw_data[0]
```

```python
plt.figure(figsize=(10, 5))
plt.plot(time, voltage, label='Raw Data', color='blue')
plt.title('Raw Data from Photodiode Area')
plt.xlabel('Time (s)')
plt.ylabel('Voltage (mV)')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plot the derivative
plt.figure(figsize=(10, 5))
plt.plot(time, np.gradient(voltage), label='Derivative', color='red')
plt.title('Derivative of Raw Data')
plt.xlabel('Time (s)')
plt.ylabel('Derivative (mV/s)')
plt.legend()
plt.grid()
plt.show()
```
